```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)
plt.title("funny line")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
